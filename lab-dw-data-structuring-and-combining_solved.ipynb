```python
# Your code goes here
import pandas as pd
```

```python
link1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
link2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
link3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
```

```python
df_1 = pd.read_csv(link1)
df_2 = pd.read_csv(link2)
df_3 = pd.read_csv(link3)
```

```python
#df_1.head(0)
#df_2.head(0)
#df_3.head(0) before
```

```python
#df_1.shape(4008, 11)
#df_2.shape(996, 11)
#df_3.shape(7070, 11)
```

```python
#Standardize name columns: lower case
df_1.columns = map(str.lower, df_1.columns)
df_2.columns = map(str.lower, df_2.columns)
df_3.columns = map(str.lower, df_3.columns)
```

```python
#Standardize name columns: white spaces

df_1.columns=df_1.columns.str.replace(" ", "_")
df_2.columns=df_2.columns.str.replace(" ", "_")
df_3.columns=df_3.columns.str.replace(" ", "_")
```

```python
#st could be replaced for state
df_1.rename(columns = {"st":"state"}, inplace = True)
df_2.rename(columns = {"st":"state"}, inplace = True)
```

```python
#df_1.head(0)
#df_2.head(0)
#df_3.head(0) after
```

```python
#to clean innecesaries spaces in case we have those
df_1.columns = df_1.columns.str.strip()
df_2.columns = df_2.columns.str.strip()
df_3.columns = df_3.columns.str.strip()
```

```python
df_concat= pd.concat([df_1, df_2, df_3], axis = 0)
```

```python
#df_concat.dropna(how = "all")
```

```python
#df_concat.gender.unique()
df_concat1 = df_concat.replace(["Male", "Femal", "female"],["M", "F","F"])
```

```python
df_concat1.gender.unique()
```

```python
df_concat1 = df_concat1.replace(["AZ", "WA", "Cali"],["Arizona", "Washington","California"])
```

```python
df_concat1.state.unique()
```

```python
df_concat1= df_concat1.replace(["Bachelors"],["Bachelor"])
```

```python

df_concat1.education.unique()
```

```python
df_concat1["customer_lifetime_value"] = df_concat1["customer_lifetime_value"].str.replace("%","")
```

```python
df_concat1 = df_concat1.replace(["Sports Car", "Luxury SUV", "Luxury Car"], ["Luxury", "Luxury","Luxury"])
```

```python
df_concat1.vehicle_class.unique()
```

```python
df_concat1["customer_lifetime_value"] = df_concat1["customer_lifetime_value"].astype('float') 
```

```python
#number of complaints
df_concat1["number_of_open_complaints"].unique()
```

```python
df_concat1 = df_concat1.replace(["1/0/00", "1/2/00", "1/1/00","1/3/00","1/5/00", "1/4/00"], ["0", "2", "1", "3", "5", "4"])
df_concat1["number_of_open_complaints"] = (df_concat1["number_of_open_complaints"].astype(float))
```

Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
round(df_concat2.groupby(["policy_type","month"])["number_of_open_complaints"].value_counts().reset_index(),2)
```
